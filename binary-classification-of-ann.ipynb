```python
df.drop(columns=["RowNumber", "CustomerId", "Surname"], inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=["Gender"],drop_first=True)
```

```python
df.head()
```

```python
df = df.replace({True: 1, False: 0})
```

```python
df.head()
```

```python
x = df.drop(columns = ['Exited'])
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit to the training data and transform it
x_train_scaled = scaler.fit_transform(x_train)

# Transform the test data using the already-fitted scaler
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
import tensorflow
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python

model.add(Dense(3,activation='sigmoid',input_dim=13))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='Adam')
```

```python
model.fit(x_train_scaled,y_train,epochs= 20)
```

```python
print(x_train_scaled.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

# Define the model with the correct input shape
model = Sequential([
    Dense(64, input_shape=(13,), activation='relu'),  # Match the feature count
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile and train the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train_scaled, y_train, epochs=100, batch_size=32,validation_split=0.2)
```

```python
model.layers[1].get_weights()
```

```python
model.predict(x_test_scaled)
```

```python
y_log = model.predict(x_test_scaled)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
